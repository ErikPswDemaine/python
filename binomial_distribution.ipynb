```python
import random
import pandas as pd

n=15
p=0.5
turns=10000
ok=0
ans=[0 for i in range(n+1)]

for i in range(turns):
    ok=0
    for j in range(n):
        if(random.random()<=p):
            ok=ok+1
    ans[ok]=ans[ok]+1
print(ans)
```

```python
S=pd.Series(ans)/turns
S.reset_index(drop=True)
```

```python
S.plot()
```
